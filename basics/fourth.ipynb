```python
# 1번 케이스에서 에러 처리를 하는 경우
error_flag = False

for i in range(10):
    # j에 대한 for Loop만 빠져나갔다.
    for j in range(10):
        if(i == 0 and j == 3):
            print("Error 발생!")
            error_flag = True
            break
        print("Item: {0}-{1}".format(i, j))
```

```python
# 1번 케이스에서 에러 처리를 하는 경우
error_flag = False

for i in range(10):
    if(error_flag):
        break
    for j in range(10):
        if(i == 0 and j == 3):
            print("Error 발생!")
            error_flag = True
            break
        print("Item: {0}-{1}".format(i, j))
```

```python
# 1번 케이스에서 에러 처리를 하는 경우
error_flag = False

for i in range(10):
    # 만약 동작하는 코드에 에러가 없다면 ?
    # 쓸대없이 코드의 성능을 저하시키는 쓸모없는 코드가 되버린다.
    if(error_flag):
        break
    for j in range(10):
        # 만약 동작하는 코드에 에러가 없다면 ?
        # 쓸대없이 코드의 성능을 저하시키는 쓸모없는 코드가 되버린다.
        if(error_flag):
            break
        for k in range(10):
            # 현재는 임의로 if문을 달아놨지만
            # 실제로는 인터럽트 같은 것으로 생각해도 된다.
            if(i == 0 and j == 3):
                print("Error 발생!")
                error_flag = True
                break
            print("Item: {0}-{1}".format(i, j))
```

```python
# 파이썬 goto의 제한은 반드시 def인 경우에만 활용할 수 있다는 것이다.
from goto import with_goto

# 루프가 너무 커지면 한계가 발생함
# 결국 C언어를 내부적으로 호출하여 처리하는 방식이 가장 선호될 것이다.
@with_goto
def goto_test():
    for i in range(10):
        for j in range(10):
            # 에러가 발생하는 시점(사실 언제 발생할지 모름)
            if (i == 3 and j == 2):
                goto .err_handler

            print("Item: {0}-{1}".format(i, j))
                
    label .err_handler
    print("Success to Error Handling!")
    print("Now check the reason!")
    
goto_test()
```
